```python
# DAY 28: THE RESURRECTION SCRIPT
# Fast Recall: Dictionaries -> DataFrames -> Cleaning -> Merging -> Analysis

import pandas as pd
import numpy as np

# 1. SETUP DATA (Dictionaries)
# Student Info
students_data = {
    'student_id': [101, 102, 103, 104, 105],
    'name': ['Ahmed', 'Bethlehem', 'Chala', 'Dawit', 'Elsa'],
    'city': ['Jimma', 'Addis Ababa', 'Adama', 'Jimma', 'Bahir Dar']
}

# Grades Data (With some missing values for us to clean)
grades_data = {
    'student_id': [101, 101, 102, 102, 103, 103, 104, 104, 105, 105],
    'subject': ['Math', 'Physics', 'Math', 'Physics', 'Math', 'Physics', 'Math', 'Physics', 'Math', 'Physics'],
    'score': [85, 90, 78, np.nan, 92, 88, np.nan, 60, 95, 98] # np.nan is missing data
}

# 2. CREATE DATAFRAMES
df_students = pd.DataFrame(students_data)
df_grades = pd.DataFrame(grades_data)

print("--- RAW GRADES (Before Cleaning) ---")
print(df_grades)
print("\n")

# 3. CLEANING (Fill Missing Scores with the Average)
# We calculate the mean of the 'score' column
mean_score = df_grades['score'].mean()
df_grades['score'] = df_grades['score'].fillna(mean_score)

print(f"--- CLEANED GRADES (Filled NaN with {mean_score:.1f}) ---")
print(df_grades)
print("\n")

# 4. MERGING (Connect Students to their Grades)
# Like SQL Join
df_merged = pd.merge(df_students, df_grades, on='student_id')

# 5. ANALYSIS (Group by Name and Calculate Average)
# Who is the top student?
student_performance = df_merged.groupby('name')['score'].mean().sort_values(ascending=False)

print("--- FINAL REPORT: STUDENT RANKING ---")
print(student_performance)
```
