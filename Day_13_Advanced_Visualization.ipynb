```python
 #Day 13 : Acadamic Challange

import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
#===============================================================================
#1.Create Students Data
#===============================================================================
data={
'Student': ['A', 'B', 'C', 'D', 'E', 'F'],
'Study_Hours': [1, 2, 3, 4, 5, 6],
'Exam_Score': [50, 55, 65, 70, 85, 95],
'Sleep_Hours': [9, 8, 7, 7, 6, 5],
}
df=pd.DataFrame(data)
display(df)
#Set Style Professional Look
sns.set_theme(style="whitegrid")
#===================================
#2.Chart:1 Line Chart(Trend)
#===================================
plt.figure(figsize=(8,4))
sns.lineplot(x='Study_Hours',y='Exam_Score',data=df,marker='o',linewidth=2.5,color='green')
plt.title('Exam Score vs Study Hours')
plt.xlabel("Hours Studied (Per Day)", fontsize=12)
plt.ylabel("Final Score (Out of 100)", fontsize=12)
plt.show( )
#=====================================
#2.Chart:1 Scatter plot(Relationship)
#=====================================
plt.figure(figsize=(8,4))
sns.scatterplot(x='Sleep_Hours',y='Exam_Score',data=df,s=100,color='blue')
plt.title('Exam Score vs Sleep Hours',fontsize=14)
plt.xlabel("Hours Slept (Per Day)", fontsize=12)
plt.ylabel("Final Score (Out of 100)", fontsize=12)
plt.show( )

#===================================
#CHART 3: Heat map (Correlation)
#===================================

corr_matrix=df[['Study_Hours','Exam_Score','Sleep_Hours']].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Matrix',fontsize=14)
plt.show()
```
