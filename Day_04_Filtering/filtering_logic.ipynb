```python
# DAY 04: FILTERING LOGIC (RE-COMMIT)
import pandas as pd

# 1. DATA
data = {
    'Name': ['Abebe', 'Kebede', 'Sara', 'Marta', 'Tola', 'Zewdu'],
    'Department': ['IT', 'Sales', 'IT', 'HR', 'Sales', 'IT'],
    'Salary': [5000, 7000, 5200, 4500, 8000, 12000],
    'Performance': [90, 60, 95, 85, 50, 98]
}
df = pd.DataFrame(data)

# 2. CHALLENGES SOLVED
# Challenge A: Sales Dept
print("--- Sales Dept ---")
display(df[df['Department'] == 'Sales'])

# Challenge B: Underperformers (< 70)
print("\n--- Underperformers ---")
display(df[df['Performance'] < 70])

# Challenge C: Low Salary OR Low Performance
print("\n--- Critical Review Needed ---")
display(df[(df['Salary'] < 5000) | (df['Performance'] < 60)])
```
