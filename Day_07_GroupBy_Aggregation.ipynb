```python
# DAY 7 CHALLENGE: SOLVED & FIXED
import pandas as pd

# 1. CREATE DATA
data = {
    "Department": ['IT', 'HR', 'IT', 'Sales', 'Sales', 'HR'],
        "Employee" : ['Abebe', 'Sara', 'Kebede', 'Marta', 'Tola', 'Zewdu'],
            "Salary" : [10000, 8000, 12000, 7000, 7500, 8200]
            }

            df = pd.DataFrame(data)

            print("--- 1. RAW DATA ---")
            display(df)

            # 2. TOTAL SALARY (SUM) - This was correct!
            salary_sum = df.groupby('Department')['Salary'].sum()

            print("\n--- 2. TOTAL SALARY PER DEPT ---")
            display(salary_sum)

            # 3. AVERAGE SALARY (MEAN) - Fixed here!
            # Added () at the end
            salary_mean = df.groupby('Department')['Salary'].mean()

            print("\n--- 3. AVERAGE SALARY PER DEPT ---")
            display(salary_mean)

            # 4. BONUS: COUNT (Number of Employees)
            employee_count = df.groupby('Department')['Employee'].count()

            print("\n--- 4. NUMBER OF EMPLOYEES ---")
            display(employee_count)
```
