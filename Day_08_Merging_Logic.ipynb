```python
import pandas as pd
orders=pd.DataFrame({'OrderID':[1, 2, 3], 'Product': ['TV', 'Phone', 'TV']
})
price=pd.DataFrame({'Product': ['TV', 'Phone', 'Laptop'], 'Price': [1000, 500, 1500] } )
df_left=pd.merge(orders,price,on='Product',how='left')
display(df_left)
total_revenue=df_left['Price'].sum()
print(total_revenue)
df_right=pd.merge(orders,price,on='Product',how='right')
print(df_right)
```
