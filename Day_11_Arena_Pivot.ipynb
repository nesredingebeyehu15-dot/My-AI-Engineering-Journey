```python
# DAY 11 ARENA: REGIONAL MANAGER REPORT
# Author: Nesredin Gebeyehu
# Description: Analysis of branch profitability and bonus eligibility calculation.

import pandas as pd
import numpy as np

# ==========================================
# 1. DATA LOADING
# ==========================================
bank_data = {
    'Branch': ['  Jemo ', 'bole', '  Jemo', 'Piassa', 'Bole', 'Megenagna'],
    'Month': ['Jan', 'Jan', 'Feb', 'Jan', 'Feb', 'Jan'],
    'Revenue': ['$10,000', '$15,000', '$12,000', '$8,000', '$14,000', '$20,000'],
    'Expenses': [5000, 8000, 6000, 7000, np.nan, 12000]
}

df_arena = pd.DataFrame(bank_data)

# ==========================================
# 2. DATA CLEANING
# ==========================================

# Standardize branch names
df_arena['Branch'] = df_arena['Branch'].str.strip().str.title()

# Clean revenue column: remove symbols and convert to float
df_arena['Revenue'] = df_arena['Revenue'].str.replace('$', '', regex=False) \
                                         .str.replace(',', '', regex=False) \
                                         .astype(float)

# Handle missing expenses by filling with mean value
mean_expense = df_arena['Expenses'].mean()
df_arena['Expenses'] = df_arena['Expenses'].fillna(mean_expense)

# ==========================================
# 3. FEATURE ENGINEERING
# ==========================================

# Calculate Net Profit
df_arena['Profit'] = df_arena['Revenue'] - df_arena['Expenses']

def check_bonus_eligibility(profit: float) -> str:
    """
    Returns 'Eligible' if profit >= 5000, else 'Not Eligible'.
    """
    if profit >= 5000:
        return "Eligible"
    else:
        return "Not Eligible"

# Apply bonus logic
df_arena['Bonus_Status'] = df_arena['Profit'].apply(check_bonus_eligibility)

print("--- PROCESSED DATA ---")
display(df_arena)

# ==========================================
# 4. REPORT GENERATION (PIVOT TABLE)
# ==========================================
# Summarize profit by Branch and Month
pivot_report = df_arena.pivot_table(
    index='Branch',
    columns='Month',
    values='Profit',
    aggfunc='sum',
    fill_value=0
)

print("--- BRANCH PROFITABILITY REPORT ---")
display(pivot_report)
```
