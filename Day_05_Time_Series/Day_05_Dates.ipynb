```python
# DAY 05: DATES & TIME SERIES (RE-COMMIT)
import pandas as pd

# 1. DATA
data = {
    'Date': ['2023-12-01', '2023-12-25', '2024-01-01', '2024-01-10'],
    'Amount': [500, 1000, 200, 300]
}
df = pd.DataFrame(data)

# 2. CONVERT TO DATETIME
df['Date'] = pd.to_datetime(df['Date'])

# 3. EXTRACT FEATURES
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

# 4. ANALYSIS
print("--- Yearly Sales ---")
display(df.groupby('Year')['Amount'].sum())
```
