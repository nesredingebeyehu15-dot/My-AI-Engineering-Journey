```python
# DAY 8 ARENA: INVENTORY MANAGEMENT SYSTEM
# Author: Nesredin Gebeyehu
# Objective: Clean, Merge, and Analyze Dirty Inventory Data

import pandas as pd
import numpy as np

# ==========================================
# 1. DATA INGESTION (RAW DATA)
# ==========================================
inventory = pd.DataFrame({
    'Item_ID': ['A001', 'A002', 'A001', 'A003', 'A004', np.nan],
    'Item_Name': [' Laptop', 'Mouse ', 'Laptop', 'Keyboard', 'Monitor', 'Cable'],
    'Stock': [10, 5, 10, 20, 0, 50]
})

prices = pd.DataFrame({
    'Item_Name': ['laptop', 'Mouse', 'KEYBOARD', 'Monitor', 'Hard Disk'],
    'Price': ['$1000', '$20', '$50', '$150', '$80']
})

print("--- 1. RAW DATA LOADED ---")
# display() is cleaner than print()
display(inventory.head(3))
display(prices.head(3))

# ==========================================
# 2. DATA CLEANING (THE JANITOR WORK)
# ==========================================

# --- A. Cleaning Inventory ---
# Step 1: Drop missing IDs
inv_clean = inventory.dropna(subset=['Item_ID']).copy()

# Step 2: Clean Item Names (CRITICAL FOR MERGE)
# አንተ 'Item_ID' ነበር ያጸዳኸው፣ ግን ለ Merge የሚያስፈልገው 'Item_Name' ነው።
inv_clean['Item_Name'] = inv_clean['Item_Name'].str.strip().str.lower()

# Step 3: Remove Duplicates
inv_clean = inv_clean.drop_duplicates()

# --- B. Cleaning Prices ---
# Step 1: Clean Price Column (Remove $ and convert to float)
prices['Price'] = prices['Price'].str.replace('$', '').astype(float)

# Step 2: Clean Item Names (CRITICAL FOR MERGE)
# Prices ላይ ያለውንም ስም ማጽዳት አለብን (KEYBOARD -> keyboard)
prices['Item_Name'] = prices['Item_Name'].str.strip().str.lower()

print("\n--- 2. DATA CLEANED & READY ---")
display(inv_clean)
display(prices)

# ==========================================
# 3. MERGING & ANALYSIS (THE INTELLIGENCE)
# ==========================================

# Step 1: Merge (Left Join to keep inventory items)
merged_df = pd.merge(inv_clean, prices, on='Item_Name', how='left')

# Step 2: Calculate Total Value (Stock * Price)
merged_df['Total_Value'] = merged_df['Stock'] * merged_df['Price']

print("\n--- 3. FINAL MERGED REPORT ---")
display(merged_df)

# Step 3: Executive Summary
total_warehouse_value = merged_df['Total_Value'].sum()

print(f"\n>>> TOTAL WAREHOUSE VALUE: ${total_warehouse_value:,.2f}")
```
